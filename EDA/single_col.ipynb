## 전체적으로 category와 integer를 나눠 분석


```python
import json

with open("./cols_name.json", "r") as f:
    dict_ = json.load(f)

cols_name_dict = dict_["columns_name"]

category_cols = dict_["category"]

integer_cols = dict_["integer"]
```

```python
len(cols_name_dict.keys()), len(category_cols), len(integer_cols)
```

```python
set(cols_name_dict.values()) - (set(category_cols).union(set(integer_cols)))
```

```python
train.rename(columns=cols_name_dict, inplace=True)

train.columns
```

```python
set_ca, set_in = set(category_cols), set(integer_cols)

str_ = "#" * 40

for col in train.columns:
    print(str_)
    print(f"col_name : {col}")
    
    if col in set_ca:
        print("value_counts")
        print(train[col].value_counts())
        
    elif col in set_in:
        print("describe")
        print(train[col].describe())
        
    print(str_, end="\n\n")
    
```
